```python
import cv2
```

```python
pathToVideo = '/Volumes/fsmresfiles/Basic_Sciences/Phys/ContractorLab/Projects/Common Files/JJM_YZ_CaImaging/2025.7/388_20250702_video_output_20min.avi'

cap = cv2.VideoCapture(pathToVideo)

fps = cap.get(cv2.CAP_PROP_FPS)
n   = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
cap.release()

print("frames:", n)
print("fps:", fps)
print("duration (s):", n/fps if fps else None)
print("duration (min):", (n/fps)/60 if fps else None)
```
